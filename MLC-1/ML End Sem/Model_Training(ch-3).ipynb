```python
from sklearn.tree import DecisionTreeClassifier
# Decision Tree is one of the Classification Model
```

```python
import pandas as pd
data = pd.read_csv("btissue.csv")
```

```python
predictors = data.iloc[:,0:9] #Segregating the predictors
predictors.head()
```

```python
target = data.iloc[:,9] #Segregating the target/class
target.head()
```

```python
from sklearn.model_selection import train_test_split
predictors_train, predictors_test, target_train, target_test = train_test_split(predictors,target,test_size=0.3,random_state=123)
#Holdout of data
```

```python
predictors_train.head()
# predictors_test.head()
# target_train.head()
# target_test.head()
```

```python
dtree_entropy = DecisionTreeClassifier(criterion='entropy', random_state = 100, max_depth=3, min_samples_leaf=5)
```

```python
# Finally the model is trained
model = dtree_entropy.fit(predictors_train,target_train)
```

Evaluating model performance

Supervised Learning - Classification

```python
prediction=model.predict(predictors_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(target_test, prediction, normalize = True)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
target_test
#len(target_test)
```

```python
confusion_matrix(target_test,prediction)
```

```python
data = pd.read_csv('auto_mpg.csv')
```

```python
data = data.dropna(axis=0, how='any')
#Remove all rows where value of any columnis ‘NaN’
```

```python
predictors = data.iloc[:,1:8]
#Seggretating the predictor variables
predictors
```

```python
target = data.iloc[:,0]
#Seggretating the target / class variable
target
```

```python
from sklearn.model_selection import train_test_split
predictors_train, predictors_test, target_train, target_test = train_test_split(predictors, target, test_size=0.3, random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
# First train model / classifier with the input dataset (training data part of it)
model = model.fit(predictors_train, target_train)
```

```python
# Make prediction using the trained model
prediction = model.predict(predictors_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(target_test, prediction)
```

```python
 r2_score(target_test, prediction)
```
