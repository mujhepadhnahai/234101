K-fold cross validation

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10,shuffle=True,random_state=123)

for fold, (train_index,test_index) in enumerate(kf.split(data),1):
    print(f"Fold {fold}")
    print(" Train indices:",train_index)
    print(" Test indices:",test_index,"\n")
```

```python
folds =[]
for train_index,test_index in kf.split(data):
    folds.append((train_index,test_index))

train_index,test_index =folds[4]

print('Train Index \n',train_index)
print('Test Index: \n',test_index)
```

```python
from sklearn.tree import DecisionTreeClassifier
predictors = data.iloc[:,0:9]
predictors.head()
```

```python
target = data.iloc[:,9]
target.head()
```

```python
from sklearn.model_selection import train_test_split
predictors_train,predictors_test,target_train,target_test = train_test_split(predictors,target,test_size=0.3,random_state=123)
```

```python
dtree_entropy = DecisionTreeClassifier(criterion="entropy",random_state=100)
model = dtree_entropy.fit(predictors_train,target_train)
prediction = model.predict(predictors_test)
from sklearn.metrics import accuracy_score
accuracy_score(target_test,prediction,normalize = True)
```

```python
from sklearn.metrics import confusion_matrix
target_test
```
