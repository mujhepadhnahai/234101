MATPLOT LIB

```python
import matplotlib.pyplot as plt
import numpy as np

date=["25/12","26/12","27/12"]
temp=[8.5,10.5,6.8]
plt.plot(date, temp)
plt.xlabel("Date") #add the Label on x-axis
plt.ylabel("Temperature") #add the Label on y-axis
plt.title("Date wise Temperature") #add the title to the chart
plt.grid(True) #add gridlines to the background

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x, y, color='red', linestyle='--', marker='o', label='sin(x)')
plt.title("Sin Graph")
plt.xlabel("x values")
plt.ylabel("y val")
plt.grid(True)
plt.show()
```

```python
y2 = np.cos(x)
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Multiple Lines')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
fig, axs = plt.subplots(2)
axs[0].plot(x,y,color = 'red',linestyle = '--')
axs[0].set_title('SIN graph')
axs[1].plot(x,z,color = 'green',linestyle='--')
axs[1].set_title('COS graph')

```

```python
import numpy as np
import matplotlib.pyplot as plt
x1=np.random.normal(0,1,1000)
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(3,2,1000)
plt.hist(x1,bins=10,color='red',edgecolor='black')
plt.hist(x2,bins=10,color='green',edgecolor='black')
plt.hist(x3,bins=10,color='blue',edgecolor='black')
plt.show()
```
