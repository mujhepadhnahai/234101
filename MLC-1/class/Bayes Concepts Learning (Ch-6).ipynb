```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_csv("apndcts.csv")
```

```python
data
```

```python
pred = data.iloc[:,0:7]
target = data.iloc[:,7]
```

```python
pred_train,pred_test, target_train, target_test = train_test_split(pred,target,test_size=0.3,random_state=123)
```

```python
gnb = GaussianNB()
```

```python
model = gnb.fit(pred_train, target_train)
```

```python
prediction = model.predict(pred_test)
```

```python
probs = model.predict_proba(pred_test)
print("Posterior probabilities: ",probs.shape)
print(probs)
print("Class Priors: ",model.class_prior_)
```

```python
accuracy = accuracy_score(target_test,prediction,normalize=True)
accuracy
```

```python
#Another problem(optional)
from sklearn import datasets,metrics

#Load sample data
iris = datasets.load_iris()
x,y = iris.data,iris.target

#Split data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=123)

#Create and train model
model = GaussianNB()
model.fit(x_train,y_train)

#Predict
y_pred = model.predict(x_test)

#Evaluate 
print("Accuracy: ",metrics.accuracy_score(y_test,y_pred))
```
