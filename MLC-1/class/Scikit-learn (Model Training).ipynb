## Supervised Learning - Classification

```python
from sklearn.tree import DecisionTreeClassifier
#Decision Tree is one of the classification model
```

```python
predictors = data.iloc[:,0:9] #Segregating the predictors
predictors.head()
```

```python
target = data.iloc[:,9] #Segregating the target/class
target.head()
```

```python
from sklearn.model_selection import train_test_split
predictors_train, predictors_test, target_train, target_test = train_test_split(predictors,target,test_size=0.3,random_state=123)
```

```python
predictors_train.head()
# predictors_test.head()
# target_train.head()
# target_test.head()
```

```python
dtree_entropy = DecisionTreeClassifier(criterion='entropy', random_state = 100, max_depth=3, min_samples_leaf=5)
```

```python
#Finally the model is trained
model = dtree_entropy.fit(predictors_train, target_train)
```

```python
prediction = model.predict(predictors_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(target_test, prediction, normalize=True)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
target_test
```

```python
confusion_matrix(target_test, prediction)
```

## Supervised Learning - Regression

```python
data = pd.read_csv('auto_mpg.csv')
data.shape
```

```python
data = data.dropna(axis=0,how='any')
#Remove all rows where value of any cols is 'NaN'
data.shape
```

```python
predictors = data.iloc[:,1:8]
#Seggregating the predictor vars
#predictors
```

```python
target= data.iloc[:,0]
#Seggregating the target/class vars
#target
```

```python
from sklearn.model_selection import train_test_split
predictors_train, predictors_test, target_train, target_test = train_test_split(predictors, target, test_size=0.3, random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
#First train model/classifier with the input dataset(training data)
model = model.fit(predictors_train, target_train)
```

```python
prediction = model.predict(predictors_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(target_test, prediction)
```

```python
r2_score(target_test, prediction)
```

## Unsupervised Learning: 

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics.cluster import v_measure_score
```

```python
data = pd.read_csv('Dataset_spine.csv')
data.head(2)
#np.shape(data)
```

```python
data_woc = data.iloc[:,0:12] #with out class var from the data set
data_woc.head(2)
```

```python
data_class = data.iloc[:,12] #Segregating the target/class var
data_class.head(2)
```

```python
f1 = data_woc['Col1'].values
#f1
```

```python
f2 = data_woc['Col5'].values
#f2
```

```python
f3 = data_woc['Col9'].values
#f3
```

```python
X = np.array(list(zip(f1,f2,f3)))
#X
```

```python
kmeans = KMeans(n_clusters=2, random_state=123)
```

```python
model = kmeans.fit(X)
```

```python
cluster_labels = kmeans.predict(X)
```

```python
v=v_measure_score(cluster_labels,data_class)
print(v)
```

### Silouette Width

```python
from sklearn.metrics import silhouette_score
```

```python
sil = silhouette_score(X, cluster_labels, metric='euclidean', sample_size=len(data))
print(sil)
#X is a feature matric for the feature subset selected for clustering
```
