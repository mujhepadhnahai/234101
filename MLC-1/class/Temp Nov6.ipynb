```python
import pandas as pd
data = pd.read_csv('btissue.csv')
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=123)
```

```python
for fold, (train_index,test_index) in enumerate(kf.split(data),1):
    print(f"Fold {fold}")
    print("Train indices: ", train_index)
    print("Test indices: ",test_index,"\n")
```

```python
data_train = data.iloc[list(kf.split(data))[2][0]]
```

```python
data_train
```

```python
data_test = data.iloc[list(kf.split(data))[2][1]]
data_test
```
