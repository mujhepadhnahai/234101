```python
#4
data =  pd.read_csv("apndcts.csv")
print(data[:5],"\n")

#5
pred = data.iloc[:,0:7]
target = data.iloc[:,7]

#6
pred_train,pred_test, target_train, target_test = train_test_split(pred,target,test_size=0.3,random_state=54)

#7
gnb = GaussianNB()
model = gnb.fit(pred_train,target_train)

#8
prediction = model.predict(pred_test)

#9
print("Prediction: ",prediction,"\n")

#10
accuracy = accuracy_score(target_test,prediction,normalize=True)
print("Accuracy: ",accuracy,"\n")

#11
probs = model.predict_proba(pred_test)
print("Posterior probs: ",probs.shape)
print(probs)
```
